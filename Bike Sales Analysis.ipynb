```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\acer.DESKTOP-1VQGBE7\OneDrive\Desktop\Power BI datasets\Sales.csv")
df
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
# Top spending customers
df.groupby(['Customer_Gender','Age_Group']).agg({'Profit':'sum'}).sort_values(by = 'Profit',ascending = False)
```

```python
# Average Order size per age group
df.groupby('Age_Group').agg({'Order_Quantity':'mean'}).sort_values(by = 'Order_Quantity')
```

```python
# Top 10 Products
df.groupby('Sub_Category').agg({'Revenue':'sum'}).sort_values(by = 'Revenue',ascending = False).head(10).plot(kind = 'pie',y = 'Revenue',figsize = (8,8))
```

```python
# Country wise revenue
df.groupby('Country').agg({'Revenue':'mean'}).plot(kind = 'bar')
```

```python
df.groupby('State').agg({'Profit':'mean'}).sort_values(by = 'Profit',ascending = False).head(5)
```

```python
order_quantity_value = df.groupby('Age_Group').agg({'Order_Quantity':'mean'}).sort_values(by = 'Order_Quantity')

print(order_quantity_value)
```

```python
# Percentage change
order_quantity_value['pct_change'] = order_quantity_value['Order_Quantity'].pct_change() * 100

print(order_quantity_value)
```
